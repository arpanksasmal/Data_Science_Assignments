```python
# 3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.
import numpy as np

class DiscreteRandomVariable:
    def __init__(self, probabilities, values):
        """
        Initializes a DiscreteRandomVariable instance.
        
        :param probabilities: List of probabilities corresponding to the values.
        :param values: List of values the random variable can take.
        """
        if len(probabilities) != len(values):
            raise ValueError("Length of probabilities and values must be the same.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("Probabilities must sum to 1.")
        
        self.probabilities = probabilities
        self.values = values

    def expected_value(self):
        """
        Calculates the expected value (mean) of the discrete random variable.
        
        :return: Expected value of the random variable.
        """
        return np.sum(np.multiply(self.values, self.probabilities))

    def variance(self):
        """
        Calculates the variance of the discrete random variable.
        
        :return: Variance of the random variable.
        """
        mean = self.expected_value()
        return np.sum(self.probabilities * (np.array(self.values) - mean) ** 2)

# Example usage
# Define values and their corresponding probabilities
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

# Create an instance of DiscreteRandomVariable
rv = DiscreteRandomVariable(probabilities, values)

# Calculate and print the expected value and variance
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
```

This Python code defines a class that models a discrete random variable, allowing the calculation of its expected value and variance. The class takes a list of possible values and their corresponding probabilities and uses them to compute these statistics.

Here's an explanation of the code:

1.Class Initialization: The DiscreteRandomVariable class is initialized with two lists: one containing the possible values the random variable can take and the other containing the corresponding probabilities. A check is performed to ensure that the lengths of these lists match and that the probabilities sum to 1, as required for a valid probability distribution.

2.Expected Value Calculation: The expected_value() method calculates the expected value (or mean) of the random variable by multiplying each value by its associated probability and summing the results. This gives a weighted average of the possible outcomes.

3.Variance Calculation: The variance() method calculates the variance of the random variable. It first calculates the expected value (mean) and then computes the weighted average of the squared deviations from the mean. This gives a measure of how spread out the values of the random variable are around the expected value.

4.Example Usage: In the example, a discrete random variable is defined with 5 possible values (1 through 5) and their corresponding probabilities. The class instance is used to calculate and display the expected value and variance based on these values and probabilities.

5.Statistical Interpretation: The expected value gives a sense of the central tendency (average outcome) of the random variable, while the variance shows the extent of variability or dispersion in the outcomes. Both are fundamental in probability theory for understanding the behavior of random variables.

```python
# 4. Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.
import numpy as np

class FairSixSidedDie:
    def __init__(self):
        self.sides = [1, 2, 3, 4, 5, 6]
        self.probabilities = [1/6] * 6  # Equal probability for each side
    
    def roll(self, n):
        """
        Simulates rolling the die `n` times.
        
        :param n: Number of rolls
        :return: List of outcomes from the rolls
        """
        return np.random.choice(self.sides, size=n, p=self.probabilities)
    
    def expected_value(self):
        """
        Calculates the expected value of the outcomes.
        
        :return: Expected value of rolling the die
        """
        return np.mean(self.sides)
    
    def variance(self):
        """
        Calculates the variance of the outcomes.
        
        :return: Variance of rolling the die
        """
        mean = self.expected_value()
        return np.mean([(side - mean) ** 2 for side in self.sides])

# Create an instance of FairSixSidedDie
die = FairSixSidedDie()

# Simulate rolling the die 1000 times
rolls = die.roll(1000)

# Calculate expected value and variance
expected_value = die.expected_value()
variance = die.variance()

# Print results
print("Expected Value of a fair six-sided die:", expected_value)
print("Variance of outcomes of a fair six-sided die:", variance)

# Optional: Print first 10 simulated outcomes for verification
print("First 10 simulated outcomes:", rolls[:10])
```

```python
# 5. Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) 
# and calculate their mean and variance.
import numpy as np
from scipy import stats

def generate_samples(distribution, *params, size=1000):
    """
    Generates random samples from a given probability distribution and calculates their mean and variance.
    
    :param distribution: The distribution function from scipy.stats (e.g., stats.binom, stats.poisson)
    :param params: Parameters for the distribution function (e.g., n, p for binomial; mu for Poisson)
    :param size: Number of samples to generate
    :return: Mean and variance of the samples
    """
    # Generate random samples
    samples = distribution.rvs(*params, size=size)
    
    # Calculate mean and variance
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

# Example usage for binomial distribution
n = 10     # Number of trials
p = 0.5    # Probability of success
binomial_mean, binomial_variance = generate_samples(stats.binom, n, p, size=1000)
print(f"Binomial Distribution: Mean = {binomial_mean}, Variance = {binomial_variance}")

# Example usage for Poisson distribution
mu = 5     # Rate (lambda) of the Poisson distribution
poisson_mean, poisson_variance = generate_samples(stats.poisson, mu, size=1000)
print(f"Poisson Distribution: Mean = {poisson_mean}, Variance = {poisson_variance}")
```

This Python code defines a function to generate random samples from various probability distributions (such as binomial or Poisson) using the scipy.stats module and calculates the mean and variance of the generated samples. The function accepts the distribution type, parameters for that distribution, and the sample size.

1.Function Definition: The generate_samples() function takes a distribution function from scipy.stats (like binomial or Poisson), the parameters for that distribution (e.g., number of trials and probability for binomial, or rate for Poisson), and the sample size.

2.Sample Generation: Using the rvs() method of the chosen distribution, the function generates random samples based on the specified parameters. This method returns a list of values sampled from the distribution.

```python
# 6. Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance,
# and standard deviation of the samples.
import numpy as np

def generate_normal_samples(mean, std_dev, size=1000):
    """
    Generates random samples from a Gaussian (normal) distribution and calculates their mean, variance, and standard deviation.
    
    :param mean: Mean of the normal distribution
    :param std_dev: Standard deviation of the normal distribution
    :param size: Number of samples to generate
    :return: Mean, variance, and standard deviation of the samples
    """
    # Generate random samples from a normal distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=size)
    
    # Calculate mean, variance, and standard deviation
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

# Example usage
mean = 0          # Mean of the normal distribution
std_dev = 1       # Standard deviation of the normal distribution
size = 1000       # Number of samples to generate

mean_val, variance_val, std_dev_val = generate_normal_samples(mean, std_dev, size)
print(f"Mean of the samples: {mean_val}")
print(f"Variance of the samples: {variance_val}")
print(f"Standard Deviation of the samples: {std_dev_val}")
```

```python
# 8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.
import numpy as np
from scipy.stats import norm

def calculate_pdf(x, mean, std_dev):
    """
    Calculates the probability density function (PDF) of a normal distribution.
    
    :param x: The point at which to evaluate the PDF
    :param mean: Mean of the normal distribution
    :param std_dev: Standard deviation of the normal distribution
    :return: Probability density function value at x
    """
    # Use scipy's norm.pdf function to calculate PDF
    pdf_value = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf_value

# Example usage
mean = 0          # Mean of the normal distribution
std_dev = 1       # Standard deviation of the normal distribution
x = 1.0           # Point at which to evaluate the PDF

pdf_value = calculate_pdf(x, mean, std_dev)
print(f"The PDF value at x = {x} for a normal distribution with mean = {mean} and std_dev = {std_dev} is: {pdf_value}")
```

```python
# 9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.
from scipy.stats import expon

def calculate_cdf(x, rate):
    """
    Calculates the cumulative distribution function (CDF) of an exponential distribution.
    
    :param x: The point at which to evaluate the CDF
    :param rate: Rate parameter (lambda) of the exponential distribution
    :return: Cumulative distribution function value at x
    """
    # Use scipy's expon.cdf function to calculate CDF
    cdf_value = expon.cdf(x, scale=1/rate)
    return cdf_value

# Example usage
rate = 1.0       # Rate parameter (lambda) of the exponential distribution
x = 2.0          # Point at which to evaluate the CDF

cdf_value = calculate_cdf(x, rate)
print(f"The CDF value at x = {x} for an exponential distribution with rate = {rate} is: {cdf_value}")
```

```python
# 15. A company is testing two different website layouts to see which one leads to higher click-through rates.
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.
# Use the following data:
'''python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

'''
import numpy as np
from scipy.stats import ttest_ind

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    """
    Perform an A/B test analysis comparing the click-through rates of two layouts.

    :param layout_a_clicks: List of clicks for Layout A
    :param layout_b_clicks: List of clicks for Layout B
    :return: t-statistic, degrees of freedom, p-value
    """
    # Convert lists to numpy arrays
    a = np.array(layout_a_clicks)
    b = np.array(layout_b_clicks)
    
    # Perform the two-sample t-test
    t_stat, p_value = ttest_ind(a, b, equal_var=False)  # Welch's t-test (unequal variances)
    
    # Calculate degrees of freedom for Welch's t-test
    n1, n2 = len(a), len(b)
    s1, s2 = np.var(a, ddof=1), np.var(b, ddof=1)
    df = ((s1/n1 + s2/n2) ** 2) / (( (s1/n1) ** 2 / (n1 - 1) ) + ( (s2/n2) ** 2 / (n2 - 1) ))
    
    return t_stat, df, p_value

# Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform the A/B test analysis
t_stat, df, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in click-through rates between the two layouts.")
else:
    print("Fail to reject the null hypothesis: No significant difference in click-through rates between the two layouts.")
```

```python
# 20. A restaurant chain collects customer satisfaction scores for two different branches. Write a program to
# analyze the scores, calculate the t-statistic, and determine if there's a statistically significant difference in
# customer satisfaction between the branches.
# Use the below data of scores:
'''python

branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]

branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

'''
import numpy as np
from scipy.stats import ttest_ind

def analyze_customer_satisfaction(branch_a_scores, branch_b_scores):
    """
    Analyze customer satisfaction scores for two different branches.

    :param branch_a_scores: List of satisfaction scores for branch A
    :param branch_b_scores: List of satisfaction scores for branch B
    :return: t-statistic, p-value
    """
    # Convert lists to numpy arrays
    branch_a = np.array(branch_a_scores)
    branch_b = np.array(branch_b_scores)
    
    # Perform the two-sample t-test
    t_stat, p_value = ttest_ind(branch_a, branch_b, equal_var=False)  # Welch's t-test (unequal variances)
    
    return t_stat, p_value

# Data
branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

# Perform the analysis
t_stat, p_value = analyze_customer_satisfaction(branch_a_scores, branch_b_scores)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a statistically significant difference in customer satisfaction between the two branches.")
else:
    print("Fail to reject the null hypothesis: No statistically significant difference in customer satisfaction between the two branches.")
```

3.Calculate T-Statistic and P-Value: The t-test calculates the t-statistic, which measures the difference between the mean scores of the two branches in relation to the variability within each branch.
```python
# 21. A political analyst wants to determine if there is a significant association between age groups and voter
# preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
# them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
# there is a significant association between age groups and voter preferences.
# Use the below code to generate data:
'''python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

'''
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate synthetic data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Statistic:", chi2_stat)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: No significant association between age groups and voter preferences.")
```

This Python code analyzes whether there's a significant association between different age groups and voter preferences using a Chi-Square test. The test checks if the distribution of voter preferences is independent of age groups.

Here's an explanation of the code:

1.Generate Data: Synthetic data for age groups and voter preferences are created. Each voter is classified into one of three age groups and one of two candidate preferences.

2.Create DataFrame: The generated data is organized into a DataFrame for easy manipulation and analysis. This DataFrame has two columns: one for age groups and one for voter preferences.

3.Create Contingency Table: A contingency table is constructed to show the frequency distribution of voter preferences across different age groups. This table is used to perform the Chi-Square test.

4.Perform Chi-Square Test: The Chi-Square test is applied to the contingency table. This test evaluates whether the observed frequencies in the table differ significantly from the frequencies expected if there were no association between age groups and voter preferences.

5.Print Results: The Chi-Square statistic, p-value, degrees of freedom, and expected frequencies are printed. The p-value indicates whether the observed distribution of preferences is significantly different from what would be expected if age groups and preferences were independent.

6.Interpret Results: The p-value is compared to a significance level of 0.05. If the p-value is less than 0.05, it suggests a significant association between age groups and voter preferences. If not, it indicates no significant association.

```python
# 24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The company wants to determine if there is a significant
# difference in customer satisfaction scores among the three product versions. They conducted a survey and collected customer satisfaction scores for
# each version from a random sample of customers. Perform an ANOVA test to determine if there is a significant difference in customer satisfaction scores.
# Use the following data:
'''python

# Sample data: Customer satisfaction scores for each product version

standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]

premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

'''
import numpy as np
from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform ANOVA test
f_stat, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

# Print the results
print("F-Statistic:", f_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("Fail to reject the null hypothesis: No significant difference in customer satisfaction scores among the product versions.")
```

This Python code assesses whether there is a significant difference in customer satisfaction scores among three different product versions (Standard, Premium, Deluxe) using an ANOVA (Analysis of Variance) test. The ANOVA test helps determine if the means of the scores for the three product versions are statistically different from each other.

Here's an explanation of the code:

1.Input Data: The code uses satisfaction scores for three product versions collected from a sample of customers. These scores are grouped into three lists corresponding to the Standard, Premium, and Deluxe versions.

2.Perform ANOVA Test: The ANOVA test is conducted to compare the mean satisfaction scores across the three product versions. This test evaluates if there are any statistically significant differences in the average scores between the groups.

3.Output Results: The ANOVA test returns:

3.1.The F-statistic, which measures the ratio of variance between the groups to the variance within the groups. A higher F-statistic indicates a larger difference between group means relative to the variability within the groups.

3.2.The p-value, which assesses the probability of observing the data if the means of all groups are equal.

4.Interpret Results: The p-value is compared to a significance level (alpha) of 0.05. If the p-value is less than 0.05, it suggests that there is a significant difference in customer satisfaction scores among the three product versions, indicating that at least one version differs significantly from the others. If the p-value is greater than or equal to 0.05, it indicates no significant difference in satisfaction scores among the versions, suggesting that the differences in scores could be due to random variation rather than the product versions.
