```python
# 1. What is the average rating for each product category?
# Remove '₹' and commas from discounted_price and convert to float
df_copy['discounted_price'] = df_copy['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# Replace '|' with '1' in the rating column
df_copy['rating'] = df_copy['rating'].str.replace('|', '1')
```

```python
# Convert the rating column to numeric
df_copy['rating'] = df_copy['rating'].astype(float)
```

```python
# Ensure there are no other unexpected characters left in the rating column
if not df_copy['rating'].between(1, 5).all():
    print("Warning: There are ratings outside the expected range of 1-5.")
```

```python
# Describe the ratings
print(df_copy['rating'].describe())
```

```python
# Calculate the average rating for each product category
df_copy['category'] = df_copy['category'].str.split('|')
df_exploded = df_copy.explode('category')
average_ratings = df_exploded.groupby('category')['rating'].mean().reset_index()
```

```python
# Display the average ratings by category
print("Average Ratings by Category:")
print(average_ratings)
```

```python
# Plot the distribution of ratings
plt.figure(figsize=(12, 8))
sns.histplot(df_copy['rating'], kde=False, color=sns.color_palette('viridis', as_cmap=True)(0.5))
plt.title("Distribution of Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.show()
```

The average rating across all categories is around 4.1, indicating that most products are well-received by customers. The distribution of ratings is generally high, with very few products rated below 4. This suggests that the majority of products meet or exceed customer expectations.

```python
# 2. What are the top rating_count products by category?
# Fill NaN values with 0
df_copy["rating_count"] = df_copy["rating_count"].fillna(0)
```

```python
# Remove commas and convert to integer
df_copy["rating_count"] = df_copy["rating_count"].astype(str).str.replace(",", "", regex=True).astype(int)
```

```python
# Ensure 'category' column is a string and not a list
df_copy["category"] = df_copy["category"].astype(str)
```

```python
# Filter out rows where rating_count is NaN
df_copy_new = df_copy[df_copy["rating_count"].notnull()]
```

```python
# Top 10 products by rating count
top_10_rating_count_products = df_copy_new[["category", "product_name", "rating_count"]].nlargest(10, "rating_count")
```

```python
# Top products by category - ensure no lists in 'category'
top_10_products_by_category = df_copy_new.groupby("category").apply(lambda x: x.nlargest(10, "rating_count")).reset_index(drop=True)
```

```python
# Plotting Top 10 products by rating count
plt.figure(figsize=(12, 8))
sns.barplot(x="rating_count", y="product_name", data=top_10_rating_count_products, palette="viridis")
plt.title("Top 10 Products by Rating Count", fontsize=16)
plt.xlabel("Rating Count", fontsize=14)
plt.ylabel("Product Name", fontsize=14)
plt.show()
```

```python
# Plotting Top Products by Category
plt.figure(figsize=(14, 10))
sns.barplot(x="rating_count", y="product_name", hue="category", data=top_10_products_by_category.head(20), palette="tab20")
plt.title("Top Products by Category", fontsize=16)
plt.xlabel("Rating Count", fontsize=14)
plt.ylabel("Product Name", fontsize=14)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

The Top 10 Products by Rating Count plot shows which products have the highest number of ratings, giving insight into the most popular products.

The Top Products by Category plot highlights the top products in each category based on their rating count, helping to identify leading products in various categories.

Both plots help in understanding which products and categories are receiving the most engagement from users.

```python
# 3. What is the distribution of discounted prices vs actual prices?
# Convert 'actual_price' to string, then remove '₹' and ',' and convert to float
df_copy['actual_price'] = df_copy['actual_price'].astype(str)
df_copy['actual_price'] = df_copy['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# Display the columns to verify
print(df_copy[["discounted_price", "actual_price"]].head())
```

```python
# Plotting the distribution of discounted prices vs. actual prices
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_copy['discounted_price'], y=df_copy['actual_price'], 
                hue=df_copy['rating'], palette="viridis", alpha=0.7)
plt.title("Distribution of Discounted Prices vs. Actual Prices", fontsize=16)
plt.xlabel("Discounted Price", fontsize=14)
plt.ylabel("Actual Price", fontsize=14)
plt.show()
```

Most products are clustered in the bottom-left to middle of the graph, indicating that the majority of products have both discounted and actual prices in the lower to mid-range.

A few products appear as outliers in the top-right corner, showing significantly higher actual prices compared to discounted prices.

The colors on the graph represent ratings, with visible points showing ratings of 3.2, 4.0, and 4.8. There are no points for ratings of 1.6 or 2.4, suggesting that products with these ratings are either absent or very rare in the dataset.

```python
# 4. How does the average discount percentage vary across categories?
df_copy['discount_percentage']
```

```python
# Preprocessing
df_copy['discount_percentage'] = df_copy['discount_percentage'].str.replace("%", "").astype(float)
```

```python
# Calculate the average discount percentage by category
average_discount_by_category = df_copy.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)
```

```python
# Print average discount percentage mean and counts
discount_percentage_mean = df_copy['discount_percentage'].mean()
more_than_average = (average_discount_by_category >= discount_percentage_mean).sum()
less_than_average = (average_discount_by_category < discount_percentage_mean).sum()

print("Average Discount Percentage Mean:", discount_percentage_mean)
print("Number of Categories with Above Average Discount:", more_than_average)
print("Number of Categories with Below Average Discount:", less_than_average)
```

The average discount percentage across all categories is approximately 47.7%.

There are 93 categories with discount percentages above this average, while 118 categories have discounts below the average.

This indicates that many categories offer discounts higher than the average, but a significant number still offer below-average discounts.

```python
# 5. What are the most popular product names?
# Extract relevant columns
new_df_prr = df_copy[["product_name", "rating", "rating_count"]]
```

```python
# Sort products by rating_count to find the most popular ones
most_popular_products = new_df_prr.sort_values(by="rating_count", ascending=False)
```

The scatter plot displays the top 20 most popular products based on the number of reviews. Each point represents a product, with the size and color indicating the number of reviews. The plot highlights the products with the highest review counts.

```python
# 6. What are the most popular product keywords?
# Concatenate product_name and about_product, then process keywords
keywords = df_copy["product_name"].str.cat(df_copy["about_product"], sep=" ", na_rep="").str.lower().str.split()
```

```python
# Flatten the list of lists into a single list
word_list = [word for sublist in keywords for word in sublist]
```

```python
# Count the frequency of each word
word_count = pd.Series(word_list)
top_keywords = word_count.value_counts().head(20)

# Print top 20 most popular product keywords
print(top_keywords)
```

```python
# 7. What are the most popular product reviews?
# Extract the top 5 popular reviews
popular_reviews = df_copy.sort_values(by='rating_count', ascending=False)[['product_name', 'about_product', 'rating_count', 'review_title', 'review_content']].head(5)
popular_reviews
```

```python
# 8. What is the correlation between discounted_price and rating?
# Clean the data
df_copy["discounted_price"] = df_copy["discounted_price"].astype(str)
df_copy["rating"] = pd.to_numeric(df_copy["rating"], errors='coerce')
new_df_dr = df_copy[["discounted_price", "rating"]]
new_df_dr["discounted_price"] = new_df_dr["discounted_price"].str.replace("₹", "").str.replace(",", "").astype(float)

# Calculate correlation
correlation = new_df_dr.corr().iloc[0, 1]
print(f"Correlation between discounted_price and rating: {correlation}")
```

```python
# Plotting the correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x="discounted_price", y="rating", data=new_df_dr, color="red", alpha=0.7)
plt.title("Correlation between Discounted Price and Rating", fontsize=16)
plt.xlabel("Discounted Price", fontsize=14)
plt.ylabel("Rating", fontsize=14)
plt.show()
```

The analysis shows a very weak positive correlation of 0.12 between rating and discounted_price. This suggests that as the rating increases, the discounted price also slightly increases. However, the correlation is not strong, indicating that other factors should be considered.

Other factors such as product quality, brand reputation, or market demand might also affect both the rating and discounted price. Therefore, it is important to consider multiple factors before making any conclusions.

# EDA of Spotify Data

## The dataset titled "Spotify Data: Popular Hip-hop Artists and Tracks" provides a curated collection of approximately 500 entries showcasing the vibrant realm of hip-hop music. These entries meticulously compile the most celebrated hip-hop tracks and artists, reflecting their significant influence on the genre's landscape. Each entry not only highlights the popularity and musical composition of the tracks but also underscores the creative prowess of the artists and their profound impact on global listeners.

```python
# 1.Load the dataframe and ensure data quality by checking for missing values and duplicate rows. 
# Handle missing values and remove duplicate rows if necessary.
# Load the DataFrame
df1 = pd.read_csv('spotify.csv')
df1.head()
```

```python
df1.shape
```

```python
df1.sample()
```

```python
df1.columns
```

```python
# Check for missing values
missing_values = df1.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

No missing values present

```python
# Check for duplicate rows
duplicate_count = df1.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
# Drop duplicate rows and reset index
df1 = df1.drop_duplicates().reset_index(drop=True)
```

```python
# Final check
print("DataFrame shape after cleaning:", df1.shape)
```

```python
# 2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.
df1['Popularity']
```

```python
# Plot the distribution of popularity
plt.figure(figsize=(10, 6))
sns.histplot(df1['Popularity'], bins=30, kde=True, color='green')
plt.title('Distribution of Track Popularity', fontsize=16)
plt.xlabel('Popularity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

The distribution of track popularity is right-skewed, indicating that most tracks have higher popularity ratings. This suggests that the dataset contains a significant number of well-liked tracks, while only a few tracks have lower popularity scores.

```python
# 3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.
# Scatter plot of Popularity vs. Duration
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df1['Duration (ms)'], y=df1['Popularity'], color='blue', alpha=0.6)

plt.title('Relationship Between Popularity and Duration of Tracks', fontsize=16)
plt.xlabel('Duration (ms)', fontsize=14)
plt.ylabel('Popularity', fontsize=14)
plt.show()
```

The scatter plot reveals that most of the tracks with higher popularity tend to have shorter durations, clustering in the top left section. A few tracks with moderate popularity have varying durations, spreading across the middle of the plot. There is one outlier with a significantly long duration but relatively lower popularity. This suggests that shorter tracks may generally be more popular, but there are exceptions, indicating that duration alone may not be the sole factor influencing popularity.

```python
# 4.Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.
# Find the artist with the highest number of tracks
artist_track_counts = df1['Artist'].value_counts()
```

```python
# 8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.
# Select numerical columns for the pair plot
numerical_df = df1[['Popularity', 'Duration (ms)']]

# Create a pair plot
sns.pairplot(numerical_df, diag_kind='kde', markers='o', hue=None)

# Display the plot
plt.suptitle('Pair Plot of Popularity and Duration', y=1.02)
plt.show()
```

Shorter tracks tend to be more popular, possibly due to their easier consumption and appeal.
Longer tracks are less likely to be popular, which could be attributed to listeners' preference for shorter, more concise music experiences.

The data indicates a trade-off between track duration and popularity. Producers and artists might need to consider this balance when creating new music to optimize for both duration and appeal.

```python
# 9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.
# Prepare data for plotting
plt.figure(figsize=(14, 8))
sns.boxplot(x='Artist', y='Duration (ms)', data=df1)
plt.xticks(rotation=90)  # Rotate artist names for better readability
plt.title('Distribution of Track Duration Across Different Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Duration (ms)', fontsize=14)
plt.show()
```

The box plot reveals significant variations in track durations across different artists. Some artists exhibit a wide range of track durations, indicating a diverse set of songs with varying lengths. For instance, artists with large interquartile ranges (IQRs) have more variability in track lengths, while those with narrow IQRs tend to have more consistent track durations. Additionally, a few artists have noticeable outliers, suggesting the presence of unusually short or long tracks compared to their usual duration. This variation in track length may reflect differing styles, genres, or creative approaches among artists.

```python
# 10.How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.
# Create a swarm plot for track popularity by artist
plt.figure(figsize=(14, 8))
sns.swarmplot(x='Artist', y='Popularity', data=df1, palette='viridis')
plt.title('Distribution of Track Popularity by Artist', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Popularity', fontsize=14)
plt.xticks(rotation=90)  # Rotate artist names for better readability
plt.show()
```

The swarm plot shows how track popularity varies for different artists. It reveals which artists have a wide range of popularity values and which ones have more consistent popularity across their tracks. For some artists, there is a broad spread of track popularity, while others have tracks that cluster around similar popularity levels.
