```python
# 1. What is the distribution of age among heart failure patients in the dataset?
plt.figure(figsize = (10, 6))
sns.histplot(df['age'], kde = True, color = 'skyblue')
plt.title('Distribution of age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

1. mean age is aprrox 60
2. most of the people is around 50-70 
3. low number of people belongs into 80 and above 

```python
# 2. How does the death rate vary with age?
plt.figure(figsize = (10, 6))
sns.kdeplot(x = df.age, hue = df.DEATH_EVENT, fill=True)
plt.title('Distribution of Age by Death Event Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

1. there are no outliers in age where the occurrence of death events deviates significantly from the overall pattern.
2. most of the death event occurs between 50-80 age groups people and low no of death events above 80.
3. most of the people suffering from heart disease, death event is significantly low for them.

```python
# 3. What is the percentge of male and female patients in the dataset?
df.sex.value_counts().plot.pie(autopct = '%1.1f%%', colors = ['lightcoral', 'lightgreen'], explode=(0, 0.1), startangle=90, labels=['Male', 'Female'])
```

To understand the ratio between male and female, lets assume 1 denotes male and 0 denotes female, then the insights will be like below:
1. there are higher number of male patients present in the dataset
2. both genders are nearly balanced but not completely.
3. the gender distribution in the dataset might introduce bias to any analyses or conclusions drawn from the data.

```python
# 4. How does the platelets count vary among different age groups?
sns.scatterplot(x=df.age, y=df.platelets)
plt.xlabel('Age')
plt.ylabel('Platelet Count')
plt.title('Relationship between Age and Platelet Count')
plt.show()

correlation = df['age'].corr(df['platelets'])
print("Correlation coefficient between age and platelet count:", correlation)
```

1. outliers are present 
2. very weak negative correlation between age and platelet count
3. avg platelet are present in 45-70 age group

```python
# 5. Is there a correlation between creatinine and sodium levels in the blood?
df[['serum_creatinine', 'serum_sodium']].corr()
```

```python
sns.heatmap(df[['serum_creatinine', 'serum_sodium']].corr(), annot= True, cmap = 'coolwarm', linewidths= .5)
```

```python
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df, hue = 'DEATH_EVENT')
plt.xlabel('Serum Creatinine')
plt.ylabel('Serum Sodium')
plt.title('Scatter Plot of Serum Creatinine vs Serum Sodium')
plt.show()
```

1. negative correlation is present.

```python
# 6. How does the prevalence of high blood pressure differ between male and female patients?
sns.countplot(x='high_blood_pressure', hue='sex', data=df, hue_order=[0, 1], palette='Set2')
plt.xlabel('High Blood Pressure')
plt.ylabel('Count')
plt.title('Count of High Blood Pressure by Gender and Death Event')
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.show()
```

1. high blood presure causes death.
2. female patients dies less than male patients due to high blood pressure.
3. 30% of the patients survived.

```python
# 7. What is the relationship between smoking habits and the occurrunce of heart failure?
df[['smoking', 'DEATH_EVENT']].corr()
```

```python
sns.heatmap(df[['smoking', 'DEATH_EVENT']].corr(), annot= True, cmap = 'coolwarm', linewidths= .5)
```

```python
sns.countplot(x='DEATH_EVENT', hue='smoking', data=df)
plt.xlabel('Death Event')
plt.ylabel('Count')
plt.title('Counts of Death Event by Smoking Habits')
plt.xticks([0, 1], ['No', 'Yes'])
plt.legend(title='Smoking', labels=['Non-Smoker', 'Smoker'])
plt.show()
```

1. very weak negeative correlation betweeen smoking habits and death events.
2. approx half of the non smokers died while approx half of the smokers survived.
3. wheather a smoker or a non smoker, any patient can face heart faliure problem.

```python
# 8. Are there any noticeable patterns in the distribution of death events across different age groups?
sns.histplot(data=df, x='age', hue='DEATH_EVENT', bins=10, multiple="stack")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Death Events Across Age Groups')
plt.show()
```

1. Among people aged between 57 and 62, as well as those aged between 68 and 73, heart attacks were the most common cause of death.
2. The age group with the highest death rate is 57-62, while the lowest death rate is observed in the 51-56 age group.
3. No data imbalance was detected.

```python
# 9. Is there any significant difference in ejection fraction between patients with and without diabetes?
sns.boxplot(x='diabetes', y='ejection_fraction', data=df)
plt.xlabel('Diabetes')
plt.ylabel('Ejection Fraction')
plt.title('Ejection Fraction by Diabetes Status')
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.show()
```

1. Outliers are present in the Ejection Fration data
2. The larger box for patients without diabetes indicates that there is more variability in ejection fraction values within this group compared to patients with diabetes.
3. Despite the potential slight shift in the central tendency towards higher values, the smaller box and the whisker line placement indicate that there is less variability in ejection fraction values among patients with diabetes.

```python
# 10. How does the serum creatinine level vary between patients who survived and those who did not?
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['DEATH_EVENT'] == 0]['serum_creatinine'], label='Survived', fill=True)
sns.kdeplot(data=df[df['DEATH_EVENT'] == 1]['serum_creatinine'], label='Not Survived', fill=True)
plt.title('Distribution of Serum Creatinine by Survival Status')
plt.xlabel('Serum Creatinine Level')
plt.ylabel('Density')
plt.legend(title='Survival Status')
plt.show()
```

1. there are outliers present in serum creatinine data.
2. Many patients survived due to low levels of serum creatinine in their bodies..
3. Higher levels of serum creatinine are associated with the occurrence of the death event. Serum creatinine levels exceeding 2 increase the probability of heart failure.
